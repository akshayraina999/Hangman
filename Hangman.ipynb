```python
import random
```

```python
name = input("Enter name:")
print("Good Luck!",name)
words = ['rainbow','college','bus','notebook']
word = random.choice(words)
print("Guess the character")
guesses = ' '
turns = 10
while turns > 0:
    failed = 0
    for char in word:
        if char in guesses:
            print(char)
        else:
            print("_")
            failed +=1
    if failed == 0:
        print("You Win")
        print("The word is", word)
        break
    guess = input("Guess character:")
    guesses = guesses+guess
    if guess not in word:
        turns -=1
        print("Wrong Guess")
        print("You have", turns,"more")
        if turns == 0:
            print("You loose")
```
